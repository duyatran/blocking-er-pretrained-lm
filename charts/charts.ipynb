```python
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import pandas as pd
# if using a Jupyter notebook, include:
%matplotlib inline
```

```python
# Structured
def plot_structured_subplots(data, ds_name, idx, fig, axes):
  ax = axes[idx]

  ax.set_title(ds_name)
  ax.set_xlabel('K')
  ax.set_ylabel('Recall')  
  
  sag_dblocker_line, = ax.plot(data.k, data.deepblocker, label='DeepBlocker')
  sag_simcse_line, = ax.plot(data.k, data.simcse, linestyle=(0, (3, 1, 1, 1)), label='SimCSE')
  sag_tsdae_line, = ax.plot(data.k, data.tsdae, label='TSDAE')

  return ax.get_legend_handles_labels()

structured_data_fns = {
  'Structured/Amazon-Google': ('./sag.csv', 'Amazon-Google (Structured)'),
  'Structured/Beer': ('./sb.csv', 'Beer (Structured)'),
  'Structured/DBLP-ACM': ('./sda.csv', 'DBLP-ACM (Structured)'),
  'Structured/Walmart-Amazon': ('./swa.csv', 'Walmart-Amazon (Structured)'),
}

plt.rc('lines', linewidth=2)
fig, axes = plt.subplots(1, 4, figsize=(14, 2.5))
handles, labels = None, None

for idx, ds in enumerate(structured_data_fns):
  input_path, ds_name = structured_data_fns[ds]
  data = pd.read_csv(input_path)
  handles, labels = plot_structured_subplots(data, ds_name, idx, fig, axes)

fig.legend(handles=handles, labels=labels, bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower center", borderaxespad=0, fancybox=True, ncol=3)
plt.tight_layout()
plt.savefig('structured.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```
